```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
# for working script in notebook
```

importing and reading Diabetes dataset using numpy and pandas library

```python
df = pd.read_csv("/content/drive/MyDrive/GC/Diabetes/diabetes.csv")
```

```python
x= np.linspace(df['Age'][10],df['Age'][20],50)
x  # giving 1D-Array.... 50 values between Age 10 index - Age 20 index.
```

```python
y = np.sin(x)
y  # giving sin values of x 
```

```python
z = np.cos(x)
z # giving cos values of value 'c'
```

```python
# plottting X, Y index using pyplot...
plt.plot(x,y)
plt.show()
```

```python
# plottting X, Z index using pyplot...
plt.plot(x,z)
plt.show()
```

```python
# Adding titles, x- and y-axis labels

plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('Sine value')
plt.title('Sinusoid wave')
plt.show()
```

```python
#parabola representation

x=np.linspace(-10,10,20)
y = x**2
plt.plot(x,y,'go--',linewidth=3, markersize=5)
plt.show()
```

**UNI-VARIATE ANALYSIS** - We use single feature for plotting.

```python
# ploting data points on Histogram (Univariate analysis)

plt.hist(df['Age'], bins=20, rwidth= .9, color='green')
plt.show()
```

```python
# plotting data points on Box plot

sns.boxplot(df['Age'], orient='h',color='yellow')
```

**BI-VARIATE ANALYSIS** -we use 2 feature for plotting

```python
# Performing Bivariate Analysis on Diabetes dataset using Scatter Plot (plotting X and Y feature)

sns.scatterplot(x = df['BloodPressure'], y=df['Glucose'], hue=df['Outcome'])
```

the graph depicts that the person who has bloddpressure almost 45 to 90 has glucose lovel almost 125-200.

**MULTI VARIATE ANALSIS** -we use any number of features

```python
## PAIR PLOTS- Useful for smaller number of features
sns.pairplot(df, hue='Outcome', size=2)  # size 2 is the size of datapoints
plt.show()
```
